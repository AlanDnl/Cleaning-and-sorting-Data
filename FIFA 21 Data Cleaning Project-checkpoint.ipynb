## Contract

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index,row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_lenght = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_lenght = end_year - start_year
        
    return start_date, end_date, contract_lenght

# Apply the fn to contract columns and create new columns
new_cols = ['Start date', 'End date', 'Contract length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i, column = new_cols[i], value = new_data[i])
    
        
```

```python
fifa[['Contract','Start date','End date','Contract length(years)']].sample(10)
```

```python
# Contract categories
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else: 
        return 'Contract'

#Add contract status column
fifa.insert(fifa.columns.get_loc('Contract length(years)')+1,'Contract Status',fifa['Contract'].apply(categorize_contract_status))
fifa.sample(5)
```

```python
fifa[['Contract','Start date', 'End date', 'Contract length(years)','Contract Status']].sample(5)
```

## Height
Players height in centimeters

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_cm(types):
    if 'cm' in types:
        total = int(types.replace('cm',''))
        return total
    else: 
        first,second = types.split("\'")
        first = int(first)
        second = int(second.replace('"',''))
        total = (first*30.48)+(second*2.54)
        return total
    
fifa['Height'] = fifa['Height'].apply(convert_cm)
```

```python
fifa['Height'] = round(fifa['Height'])
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns = {'Height':'Height(cm)'})
fifa.sample(1)
```

## Weight
Player's wieght in kg

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_to_kg(convert):
    if 'kg' in convert:
        total = int(convert.replace('kg',''))
        return total
    else: 
        pounds = int(convert.replace('lbs',''))
        total = round(pounds*0.453)
        return total
    
fifa['Weight'] = fifa['Weight'].apply(convert_to_kg)
fifa.sample(3)
```

```python
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns ={'Weight':'Weight(kg)'})
fifa.sample(1)
```

## Hits
Number of times the player has been searched for in the FIFA database

```python
 fifa['Hits'].dtype
```

```python
fifa['Hits'].unique()
```

```python
# Función para limpiar la columna 'Hits'
def convert_K(valor):
    if pd.notna(valor):
        if 'K' in str(valor):
            # Si hay una 'K', multiplicar por 1000
            return int(float(str(valor).replace('K', '')) * 1000)
        else:
            # Si no hay 'K', convertir a entero
            return int(float(valor))
    else:
        return np.nan

fifa['Hits'] = fifa['Hits'].apply(convert_K)
fifa.sample(5)
```
